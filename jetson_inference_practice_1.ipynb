```python
import jetson.inference
import jetson.utils
```

```python
#Load image
image = jetson.utils.loadImage('shiba.jpg')
print(image)
```

```python
import matplotlib.pyplot as plt
```

```python
image_np = jetson.utils.cudaToNumpy(image)
print(image_np.shape)
```

```python
plt.imshow(image_np)
plt.show()
```

```python
#Load the recognition network (default - GoogleNet)
net = jetson.inference.imageNet()
```

```python
#Classify the image
class_idx, confidence = net.Classify(image)
```

```python
print(class_idx, confidence)
```

```python
class_desc = net.GetClassDesc(class_idx)
print(class_desc)
```
